```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_base = pd.read_csv("category_data.csv")
```

```python
df = df_base.copy()
```

```python
df.head()
```

```python
df.head(30)
```

```python
# Yl ve kategoriye gre 'support vector classification_r' ortalamasn aln
pivot_df = df.pivot_table(index='Category', columns='Year', values='support vector classification_r')

# s haritas oluturun
plt.figure(figsize=(10,8))
sns.heatmap(pivot_df, annot=True, cmap='coolwarm')

plt.title('Kategorilere Gre Yllk Support Vector Classification R Skorlar')
plt.show()
```

```python
df = df[(df['Year'] >= 2014) & (df['Year'] <= 2022)]
plt.figure(figsize=(12,8))
#sns.lineplot(data=df, x='Year', y='support vector classification_r', hue='Category')
line_plot = sns.lineplot(data=df, x='Year', y='support vector classification_r', hue='Category')
plt.title('Sentiment Scores for Categories')
plt.xlabel('Year')
plt.ylabel('Support Vector Classification Average Scores')
plt.legend(loc='lower right', bbox_to_anchor=(1.25, -0.30), ncol=2)
plt.show()
```
